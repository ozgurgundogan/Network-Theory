```python
# Most important 10 locations has been highlighted for Piazza Mazini Square in Rome.
# Here, Katz Centrality Metric Has been used.
```

```python
import numpy as np
import networkx as nx
import scipy.io
import matplotlib.pyplot as plt
```

```python
#load adjancecy matrix
aa=scipy.io.loadmat('adj.mat')
```

```python
mat=aa['A']
```

```python
# convert adjancency matrix to graph
G=nx.from_numpy_matrix(mat)
```

```python
sortedlist=sorted(enumerate([val for key,val in nx.current_flow_betweenness_centrality(G).items()]),key=lambda x:x[1],reverse=True)
```

```python
fig = plt.figure(1,figsize=(12,8))
plt.clf()

idx=[k[0] for k in sortedlist[:10]]
color=[[0,0,1]]*mat.shape[0]
size=[[15]]*mat.shape[0]
for el in idx:
    color[el]=[1,0,0]
    size[el]=[30]
nx.draw(G, pos=nx.spring_layout(G),node_size=size,node_color=color)    
fig.savefig('flow_betweenness_centrality.png', dpi=300)
```
